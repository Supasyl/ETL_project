```python
import pandas as pd
import requests
import json
```

#### Create an URL  for our API, returning all information for lego sets produced in 2020 in JSON format


```python
url = "https://brickset.com/api/v3.asmx/getSets?apiKey=3-DLfb-T3ZA-qQLjW&userHash=&params={'year':'2020','pageSize':900}"

response = requests.request("GET", url)
response.json()
```

### Load our Results into a JSON file

```python
with open('sets2020.json', 'w') as fp:
    json.dump(response.json(), fp)
```

#### Open our JSON file and Create the desired dataframe

```python
 with open("sets2020.json","r") as fp:
    json_d = json.load(fp)
    
sets = json_d['sets']
sets_df = pd.DataFrame(sets)
```
